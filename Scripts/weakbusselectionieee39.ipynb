```python
skss = pd.read_excel(r"C:\Users\PF-WS-008\Desktop\Tesis\Results IEEE39\Base case\Weak bus selection\ShortCircuitCalculation.xlsx",
                     skiprows=12, header = 0)
skss.head()
```

```python
df_bus = skss[skss["Terminal"].str.contains("Bus", na = False)]
df_bus.head()
print(len(df_bus))
```

```python
df_bus.columns
```

```python
buses_sorted = df_bus.sort_values('Sk" \n (MVA)', ascending=True )
buses_sorted = buses_sorted.rename(columns = {'Sk" \n (MVA)':"Skss_MVA"})
buses_sorted.head(5)
```

```python
import matplotlib.pyplot as plt
import os
resultspath = r"C:\Users\PF-WS-008\Desktop\Tesis\Results IEEE39\Base case\Weak bus selection"
outfile = os.path.join(resultspath,"Weakest_bus_Skss.png")

plt.figure(figsize = (10,4))
plt.grid(axis="y",linestyle = "--",alpha = 0.6)
plt.ylim(0, buses_sorted["Skss_MVA"].max() * 1.1)
plt.yticks(range(0, int(buses_sorted["Skss_MVA"].max() * 1.1) + 1, 5000))
plt.bar(buses_sorted["Terminal"], buses_sorted['Skss_MVA'])
plt.xticks(rotation=45)
plt.ylabel("Short-circuit power Sk'' (MVA)")
plt.xlabel("Bus")
plt.title("Weakest bus by short-circuit power")
plt.tight_layout()
plt.savefig(outfile,dpi = 300,bbox_inches = "tight")
plt.close()
```
