```python
path = r"C:\Users\UI450907\Desktop\TE RWEST\Tesis\Results _IEEE39\SCR\Skss Study.xlsx"

dfs = pd.read_excel(path,sheet_name=None,skiprows=11)
for name,df in dfs.items():
    new_cols = df.iloc[0].astype(str)
    df = df.iloc[1:].reset_index(drop = True)
    df.columns = new_cols
    df = df.rename(columns={'Sk" \n (MVA)':"Skss"})
    dfs[name] = df
print(dfs.keys())
```
